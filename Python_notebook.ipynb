```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r'C:\Users\USER\Downloads\archive (1)\bank_fraud.csv')

print(df.shape)
print(df.head())
```

```python
# Overview - Fraud vs Legitimate
fraud_counts = df['is_fraud'].value_counts()
fraud_counts.index = ['Legitimate', 'Fraud']

print("Transaction Overview:")
print(f"Total Transactions: {len(df):,}")
print(f"Total Fraud Cases: {df['is_fraud'].sum():,}")
print(f"Fraud Rate: {df['is_fraud'].mean()*100:.2f}%")

# Donut Chart
plt.figure(figsize=(6,4))
plt.pie(fraud_counts, labels=fraud_counts.index, autopct='%1.2f%%', startangle=90, wedgeprops=dict(width=0.5))
plt.title('Fraud vs Legitimate Transactions')
plt.show()
```

```python
# Fraud by Country
fraud_country = df[df['is_fraud'] == 1]['country'].value_counts().head(10)

plt.figure(figsize=(6,4))
sns.barplot(x=fraud_country.values, y=fraud_country.index, palette='Reds_r')
plt.title('Top 10 Countries by Fraud Cases')
plt.xlabel('Total Fraud Cases')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

```python
fraud_merchant = df[df['is_fraud'] == 1]['merchant_category'].value_counts()

plt.figure(figsize=(6,4))
sns.barplot(x=fraud_merchant.index, y=fraud_merchant.values, palette='Blues_r')
plt.title('Fraud by Merchant Category')
plt.xlabel('Merchant Category')
plt.ylabel('Total Fraud Cases')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
fraud_payment = df[df['is_fraud'] == 1]['payment_method'].value_counts()

plt.figure(figsize=(6,4))
sns.barplot(x=fraud_payment.values, y=fraud_payment.index, palette='Oranges_r')
plt.title('Fraud by Payment Method')
plt.xlabel('Total Fraud Cases')
plt.ylabel('Payment Method')
plt.tight_layout()
plt.show()
```

```python
fraud_hour = df[df['is_fraud'] == 1]['hour_of_day'].value_counts().sort_index()

plt.figure(figsize=(6,4))
sns.lineplot(x=fraud_hour.index, y=fraud_hour.values, color='red', linewidth=2)
plt.title('Fraud by Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Total Fraud Cases')
plt.xticks(range(0,24))
plt.tight_layout()
plt.show()
```

```python
fraud_device = df[df['is_fraud'] == 1]['device_type'].value_counts()

plt.figure(figsize=(6,4))
sns.barplot(x=fraud_device.values, y=fraud_device.index, palette='Purples_r')
plt.title('Fraud by Device Type')
plt.xlabel('Total Fraud Cases')
plt.ylabel('Device Type')
plt.tight_layout()
plt.show()
```
